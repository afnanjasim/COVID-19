```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Setting up our axis
a = np.arange(-1, 1, 0.001)
b = np.arange(-1, 1, 0.001)
```

```python
def f_nonzero(x,y):
    return  (x * y * (x**2 - y**2))/(x**2 + y**2)
```

```python
#Creating a mesh for plotting    
(A, B) = np.meshgrid(a, b)

#Calculating Values for Z Axis
z_nonzero = f_nonzero(A, B)
```

```python
#Plotting Initialization
fig = plt.figure()
fig.set_size_inches((10, 8.5))
ax = plt.axes(projection='3d')

ax.plot_wireframe(A, B, z_nonzero, cmap='twilight_r', alpha=0.92)     #Plotting Function's value at (x,y)!=(0,0) 
ax.view_init(50, 50)
plt.show()
```

```python
dx = dy = 0.0001

#Defining the two partial derivatives
def p_df_dx(x, y):
    return (f_nonzero(x+dx, y)-f_nonzero(x,y))/dx

def p_df_dy(x, y):
    return (f_nonzero(x, y+dy)-f_nonzero(x,y))/dy

#Calculating the two partial derivatives
pd_wrt_x = p_df_dx(A, B)
pd_wrt_y = p_df_dy(A, B)
```

```python
%matplotlib inline
#Plottting Initialization
fig, (ax1, ax2) = plt.subplots(1,2, subplot_kw=dict(projection='3d'))
fig.set_size_inches((16, 7))

#Plotting Partial Derivative wrt to X 
ax1.plot_surface(A, B, pd_wrt_x, cmap='plasma_r')
ax1.set_title("Partial Derivatives wrt X")
ax1.view_init(20, 80)

#Plotting Partial Derivative wrt to Y
ax2.plot_surface(A, B, pd_wrt_y, cmap='plasma_r', alpha=0.95)
ax2.view_init(20, 80)
ax2.set_title("Partial Derivatives wrt Y")

plt.show()
```
