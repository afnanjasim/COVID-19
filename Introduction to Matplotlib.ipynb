```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = 12, 8
plt.rcParams.update({'font.size': 20})
```

```python
x1 = np.array([0, 1, 2, 3, 4, 5, 6])
y1 = np.array([0, 1, 4, 9, 16, 25, 36])

plt.plot(x1, y1)
```

```python
x2 = np.array([0, 1, 2, 3, 4, 5, 6])
y2 = np.pi * x2

plt.plot(x2, y2)
```

```python
plt.scatter(x1, y1)
plt.scatter(x2, y2)
```

```python
x3 = np.array(np.arange(-2 * np.pi, 2 * np.pi , 0.01))
y3 = np.sin(x3)
y4 = np.cos(x3)

plt.plot(x3, y3)
```

```python
plt.plot(x3, y3)
plt.show()
plt.plot(x3, y4)
plt.show()
```

```python
def kinematics(t, a):

    v = a * t
    s = 0.5 * a * t ** 2
    
    return v, s

t = np.arange(0, 10, 0.1)

v1 = kinematics(t, 9.8)[0]
v2 = kinematics(t, 1.6)[0]

s1 = kinematics(t, 9.8)[1]
s2 = kinematics(t, 1.6)[1]
```

```python
plt.plot(t, v1, label="Earth")
plt.plot(t, v2, label="Moon")

plt.xlabel("Time (t)")
plt.ylabel("Velocity (v)")

plt.title("Velocity Graph")

plt.legend()
#plt.grid()
```

```python
plt.plot(t, s1, 'r--', label="Earth")
plt.plot(t, s2, 'g-o', label="Moon")

plt.xlabel("Time (t)")
plt.ylabel("Position (s)")

plt.title("Position Graph")

plt.legend()
plt.grid()

plt.xlim(0, 2)
plt.ylim(0, 30)

plt.savefig("Position")
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)

fig.set_size_inches((16, 7))
plt.suptitle("Falling Bodies on Earth")

ax1.plot(t, v1)
ax1.set_xlabel("Time")
ax1.set_ylabel("Velocity")


ax2.plot(t, s1)
ax2.set_xlabel("Time")
ax2.set_ylabel("Position")
```
