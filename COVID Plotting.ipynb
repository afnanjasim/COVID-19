```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['figure.figsize'] = 20, 10
```

```python
data = pd.read_csv('covidData.csv')
date, total, death, new = data['date'], data['total_cases'], data['total_deaths'], data['new_cases']

totalIND = total[10606:10779]
dateIND = date[10606:10779]
deathIND = death[10606:10779]
newIND = new[10606:10779]
```

```python
data.set_index("location", inplace=True)
data.head()
china = data.loc[['China']]
```

```python
totalCHN = china['total_cases']
dateCHN = china['date']
deathCHN = china['total_deaths']
newCHN = china['new_cases']
```

```python
fig, ax = plt.subplots(1,1)

ax.scatter(dateIND, totalIND)
ax.scatter(dateCHN, totalCHN)

ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.grid()
ax.set_title('Total COVID Cases')
```

```python
fig, (ax1, ax2) = plt.subplots(2,1)

ax1.semilogy(dateIND, totalIND)
ax1.xaxis.set_major_locator(plt.MaxNLocator(15))

ax2.plot(dateIND, totalIND)
ax2.xaxis.set_major_locator(plt.MaxNLocator(15))
```

```python
fig, (ax1, ax2) = plt.subplots(2,1)

ax1.plot(dateIND, newIND, 'o-')
ax1.set_title("New Cases In India")

ax2.plot(dateCHN, newCHN, 'o-')
ax2.set_title("New Cases In China")

ax1.xaxis.set_major_locator(plt.MaxNLocator(15))
ax2.xaxis.set_major_locator(plt.MaxNLocator(15))

plt.show()
```
